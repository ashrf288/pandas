## For the top-selling game of all time, how many standard deviations above/below the mean are its sales for North America?

```python
standard_deviation = data['NA_Sales'].std()
mean = data['NA_Sales'].mean()
top_seles = data['NA_Sales'].iloc[0]
sales_in_range=(top_seles-mean) / standard_deviation
sales_in_range
```

## The Nintendo Wii seems to have outdone itself with games. How does its average number of sales compare with all of the other platforms?

```python

wii_sales = df[ df['Platform']=='Wii' ].Global_Sales
other_platforms = df[ df['Platform']!='Wii' ].Global_Sales
nentendo_outdone=wii_sales.mean()-other_platforms.mean()
nentendo_outdone
```
